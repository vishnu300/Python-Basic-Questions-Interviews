```python
import streamlit as st
import PyPDF2  # For PDF manipulation
import io  # For handling file-like objects
```

```python
def compress_pdf(input_file, output_file):
    with open(input_file, 'rb') as in_file, open(output_file, 'wb') as out_file:
        reader = PyPDF2.PdfReader(in_file)
        writer = PyPDF2.PdfWriter()

        # You can add compression logic here, e.g.,
        # Adjust image compression, font subsetting, etc.

        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            writer.add_page(page)

        writer.write(out_file)
```

```python
def main():
    st.title("PDF Compressor")

    uploaded_file = st.file_uploader("Choose a PDF file", type="pdf")

    if uploaded_file is not None:
        bytes_data = uploaded_file.read()
        pdf_reader = PyPDF2.PdfReader(io.BytesIO(bytes_data))
        original_filename = uploaded_file.name

        # Create a temporary file
        with open("temp.pdf", "wb") as f:
            f.write(bytes_data)

        # Compress the PDF
        compressed_filename = "compressed_" + original_filename
        compress_pdf("temp.pdf", compressed_filename)

        # Read the compressed file
        with open(compressed_filename, "rb") as f:
            compressed_pdf = f.read()

        # Display download button
        st.download_button(
            label="Download Compressed PDF",
            data=compressed_pdf,
            file_name=compressed_filename,
            mime='application/pdf'
        )


main()
```
